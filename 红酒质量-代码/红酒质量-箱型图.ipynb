```python
df = pd.read_csv('E:\jupyter\机器学习项目\红酒质量-svm\winequality-red.csv')
```

```python
OrderedCols = df.columns.values
```

```python
sns.set_style('darkgrid')
colors = ['#0055ff', '#ff7000', '#23bf00']
CustomPalette = sns.set_palette(sns.color_palette(colors))
```

```python
# 创建子图
fig, ax = plt.subplots(3, 4, figsize=(15,7), dpi=100)
# 绘制箱线图
# x,y 判断对应子图的位置
# 如果 i < 3 ,指的是前面几个数值型的， 箱线图描述的是该列所有数值的分布情况
# 如果 i > 3 ,指的是后面几个非数值型的 , 箱线图描述的是每一列中不同种类与售价的分布情况
for i, col in enumerate(OrderedCols[:-1]):
    x = i // 4
    y = i % 4
    sns.boxplot(data=df, y=col, ax=ax[x, y])
    ax[x, y].yaxis.label.set_size(15)

ax[2, 3].axis('off')
plt.tight_layout()    
plt.savefig("红酒质量-箱型图.png", dpi=300) 
plt.show()
```
